```python
import pandas as pd
df=pd.read_csv('house_lime.csv')
```

```python
contribution_cols=[x for x in df.columns if 'contribution' in x]
```

```python
# copy the data
df_min_max_scaled = df.copy()
# apply normalization techniques
for column in contribution_cols:
	df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())	
```

```python
df_min_max_scaled[contribution_cols]
```

```python
df_min_max_scaled.to_csv("house_lime.csv",index=None)
```
