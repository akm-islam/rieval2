# Lime Data

```python
orig_data=pd.read_csv("../../../threepublisher.csv")
orig_data['myindex']=orig_data.index
```

```python

contribution_columns=list(filter(lambda x: "_contribution" in x, df.columns))
```

```python
feature_columns=[x.replace("_contribution","") for x in contribution_columns]
```

```python
feature_data=orig_data[feature_columns]
mart_data=feature_data.copy()
mart_data['model']='MART'
random_for=feature_data.copy()
random_for['model']='RandomFor'
```

```python
merged_data=pd.concat([mart_data,random_for])
```

```python
merged_data.reset_index(inplace=True, drop=True)
```

```python
merge_final=pd.concat([merged_data,df],axis=1)
```

```python
merge_final
```

```python
mycols=np.array(['1-qid','two_realRank','two_realRank2','State','predicted','year','model',"publisher"]) # This columns must exist
for col in mycols:
    if(col not in merge_final.columns):
        print(col)
```

```python
temp=np.concatenate((contribution_columns,feature_columns,mycols))
```

```python
threepublisher_lime=merge_final[temp]
```

```python
threepublisher_lime[contribution_columns]=threepublisher_lime[contribution_columns].apply(lambda x:round(x,3))
```

```python
threepublisher_lime.shape
```

```python
threepublisher_lime.to_csv("threepublisher_lime2.csv",index=None)
```
