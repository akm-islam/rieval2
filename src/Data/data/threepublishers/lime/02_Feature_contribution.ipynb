```python
import pandas as pd
df=pd.read_csv('./threepublisherlime.csv')
import numpy as np
```

# Rank Data

```python
rank_columns=list(filter(lambda x: "_feature_rank" in x, df.columns))
print(len(rank_columns))
```

```python
mycols=np.array(['1-qid','two_realRank','two_realRank2','State','predicted','year','model']) # This columns must exist
for col in mycols:
    if(col not in df.columns):
        print(col)
```

```python
temp=np.concatenate((rank_columns,mycols))
```

```python
threepublisher_rank=df[temp]
```

```python
threepublisher_rank.to_csv("threepublisher_rank.csv",index=None)
```

# Lime Data

```python

contribution_columns=list(filter(lambda x: "_contribution" in x, df.columns))
len(contribution_columns)
```

```python
mycols=np.array(['1-qid','two_realRank','two_realRank2','State','predicted','year','model']) # This columns must exist
for col in mycols:
    if(col not in df.columns):
        print(col)
```
